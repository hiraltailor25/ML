```python
#Genrate some data
categories = np.array(['A','B','C','D'])
values1 = np.array([3,8,1,10])
values2 = np.array([6,2,7,5])

#Set the width of the bars
bar_width = 0.35

#Create the plot
plt.bar(categories, values1, bar_width, label='Dataset 1')
plt.bar(categories, values2, bar_width+0.20, label='Dataset 2')

#Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Grouped Bar Chart Example')

#Add legend
plt.legend()

#Display the plot
plt.show()
```

```python
#Genrate some data
categories = np.array(['A','B','C','D'])
values1 = np.array([3,8,1,10])
values2 = np.array([6,2,7,5])
values3 = np.array([2,5,6,1])

#Set the width of the bars
bar_width = 0.35

#Create the plot
plt.bar(categories, values1, bar_width, label='Dataset 1')
plt.bar(categories, values2, bar_width+0.20, label='Dataset 2')
plt.bar(categories, values3, bar_width-0.20, label='Dataset 3')

#Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Grouped Bar Chart Example')

#Add legend
plt.legend()

#Display the plot
plt.show()
```

```python
#Genrate some data
categories = np.array(['A','B','C','D'])
values = np.array([3,8,1,10])

#Create the plot
plt.pie(values, labels=categories)

#Add title
plt.title('Simple Pie Chart Example')

#Display the plot
plt.show()
```

```python
#Genrate some data
categories = np.array(['A','B','C','D'])
values = np.array([3,8,1,10])

#Define which  slice to explode
explode=(0.1, 0, 0.1, 0)

#Create the plot
plt.pie(values, labels=categories, explode=explode)

#Add title
plt.title('Exploded Pie Chart Example')

#Display the plot
plt.show()
```

```python
#Genrate some data
categories = np.array(['A','B','C','D'])
values = np.array([3,8,1,10])

#Define which  slice to explode
explode=(0.1, 0, 0.1, 0)

#Create the plot
plt.pie(values, labels=categories, explode=explode, autopct='%1.1f%%')

#Add title
plt.title('Exploded Pie Chart Example')

#Display the plot
plt.show()
```
