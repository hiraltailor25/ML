```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b--')
plt.show()
```

```python
plt.plot(x,y, linewidth=2.0)
plt.show()
```

```python
plt.plot(x,y, marker='+')
plt.show()
```

```python
t=np.arange(0.,5.,0.2)

#red dashes blue squares and green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```
